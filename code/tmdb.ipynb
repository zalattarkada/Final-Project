```python
df = pd.DataFrame(columns=["IMDb_ID", "Year", "Title","Genres","Production Companies", "Director", "Cast", "Budget", "Revenue", "Rating"])

base_url = "https://api.themoviedb.org/3"  # Replace with your base url
api_key = "5c419f6d35427c1d1d7a6145167b0d2f"  # Replace with your API key

for year in range(2000, 2024):
    movies_count = 0
    page = 1
    while movies_count < 100:
        discover_url = f"{base_url}/discover/movie?api_key={api_key}&language=en-US&sort_by=revenue.desc&include_adult=false&include_video=false&page={page}&primary_release_year={year}"
        response = requests.get(discover_url).json()

        # Check if there are no more movies
        if not response["results"]:
            break

        movies = response["results"]
        for movie in movies:
            movie_id = movie["id"]

            # Fetch movie details
            details_url = f"{base_url}/movie/{movie_id}?api_key={api_key}&language=en-US"
            details_response = requests.get(details_url).json()

            # Fetch movie credits
            credits_url = f"{base_url}/movie/{movie_id}/credits?api_key={api_key}"
            credits_response = requests.get(credits_url).json()

            # Get the list of genres
            genres = [genre["name"] for genre in details_response["genres"]]

            # Get the director and cast
            director = ""
            cast = []
            for person in credits_response["crew"]:
                if person["job"] == "Director":
                    director = person["name"]
            for person in credits_response["cast"]:
                cast.concat(person["name"])
            
            # Fetch IMDb ID based on TMDB movie ID
            external_id_url = f"{base_url}/movie/{movie_id}/external_ids?api_key={api_key}"
            external_id_response = requests.get(external_id_url).json()
            imdb_id = external_id_response.get("imdb_id")

            # Get the list of production companies
            production_companies = [company["name"] for company in details_response["production_companies"]]

            # Add the movie data to the dataframe
            df = df.append({"Year": year,
                            "Title": movie["title"],
                            "Genres": genres,
                            "Director": director,
                            "Cast": cast,
                            "IMDb_ID": imdb_id,
                            "Production Companies": production_companies,
                            "Budget": details_response["budget"],
                            "Revenue": details_response["revenue"],
                            "Rating": details_response["vote_average"]}, ignore_index=True)

            # Increment the count of movies
            movies_count += 1

            # If already got 100 movies, break the loop
            if movies_count >= 100:
                break

        # Move to the next page
        page += 1

        # Sleep to avoid hitting rate limit
        time.sleep(1)

print(df)
```
