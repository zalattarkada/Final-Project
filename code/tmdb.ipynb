```python
import requests
import json
import pandas as pd

api_key = "5c419f6d35427c1d1d7a6145167b0d2f"  # replace with your actual API key

# Define the base URL for the Discover API
discover_url = "https://api.themoviedb.org/3/discover/movie"
movie_url = "https://api.themoviedb.org/3/movie/{movie_id}"  # for fetching individual movie info

# Prepare an empty list to store the results
top_grossing_movies = []

# Iterate over the years 1920-2023
for year in range(1940, 2024):
    print(f"Processing year {year}...")

    # Define the parameters for the API request
    params = {
        "api_key": api_key,
        "language": "en-US",
        "sort_by": "popularity.desc",  # sort by popularity in descending order
        "primary_release_year": year,  # filter by release year
        "page": 1  # start with the first page
    }

    # Send the API request
    response = requests.get(discover_url, params=params)
    data = response.json()

    # Add the results to the list
    top_movies = data["results"]

    # Continue to fetch additional pages until we have 200 movies or all pages have been fetched
    while len(top_movies) < 20 and params["page"] < data["total_pages"]:
        params["page"] += 1  # increment the page number
        response = requests.get(discover_url, params=params)
        data = response.json()
        top_movies.extend(data["results"])

    # 200 mejores
    top_movies = top_movies[:20]

    for movie in top_movies:
        # Get additional movie details
        movie_params = {
            "api_key": api_key,
            "language": "en-US",
            "append_to_response": "credits",  # include credits information
        }
        movie_response = requests.get(movie_url.format(movie_id=movie["id"]), params=movie_params)
        movie_data = movie_response.json()
        
        # Extract the required information from the movie_data
        movie["budget"] = movie_data.get("budget", "N/A")
        movie["director"] = movie_data["credits"]["crew"][0]["name"] if movie_data["credits"]["crew"] else "N/A"
        movie["actors"] = [actor["name"] for actor in movie_data["credits"]["cast"][:3]] if movie_data["credits"]["cast"] else "N/A"
        movie["genres"] = [genre["name"] for genre in movie_data["genres"]] if movie_data["genres"] else "N/A"

    # Add the movies to the list
    top_grossing_movies.extend(top_movies)

# Create a pandas DataFrame from the movie information
df = pd.DataFrame(top_grossing_movies)

# Select the desired columns
df = df[["title", "release_date", "budget", "revenue", "director", "actors", "genres"]]

# Print the DataFrame
print(df)
```
